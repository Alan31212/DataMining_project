```python
#Checking the data type and missing value
import pandas as pd

df=pd.read_csv('HW1.csv', header=0)
cols=df.columns
print(type(df))
print('\nColumnName, DataType, MissingValues')
for i in cols:
    print(i, ',', df[i].dtype,',',df[i].isnull().any())
```

```python
#quention b
from statistics import mean 
print((df.Income-min(df.Income))/(max(df.Income)-min(df.Income))*(5-1)+1)
```

```python
#quention c
print(mean(df.Age))
std = df["Age"].std()
print(std)
print((df.Age-mean(df.Age))/std)
```

```python
#quention d
def Age_categorize(Age):
    if Age >=1 and Age <=20:
        return 'Young'
    elif Age >= 21 and Age <= 40:
        return 'MidAge'
    else:
        return 'Old'

df['Age_categorize']= df['Age'].apply(Age_categorize)
print(df[['Age', 'Age_categorize']])
```

```python
#quention e
def Gender_catagory(Gender):
    if Gender == 'M':
        return '0'
    else:
        return '1'

def Incidental_Catagory(Incidentals):
    if Incidentals == 'Yes':
        return '0'
    else:
        return '1'
    
def Genre_catagory(Genre):
    if Genre == 'Action':
        return '1'
    elif Genre == 'Drama':
        return '2'
    else:
        return '3'

df['Gender_catagory']= df['Gender'].apply(Gender_catagory)
df['Incidental_Catagory']= df['Incidentals'].apply(Incidental_Catagory)
df['Genre_catagory']= df['Genre'].apply(Genre_catagory)
print(df[['Gender','Gender_catagory','Incidentals','Incidental_Catagory','Genre', 'Genre_catagory']])
```

```python
#quention Q3(2)
correlations = df.corr()
print(correlations)
#Age and income have a strong correlation with rentals.
```

```python
#quention Q3(3)
from scipy.stats import chi2_contingency
from scipy.stats import f_oneway

categorical_vars = ['Gender', 'Incidentals']
for var in categorical_vars:
    contingency = pd.crosstab(df['Genre'], df[var])
    chi2, p, _, _ = chi2_contingency(contingency)
    print(f"Genre and {var}: p-value= {p}")

variables = ['Income', 'Age', 'Rentals', 'Avg Per Visit']
for var in variables:
    groups = [df[df['Genre'] == g][var] for g in df['Genre'].unique()]
    f_statistic, p_value = f_oneway(*groups)
    print(f"Genre and {var}: p-value = {p_value}")
# if p-value is less than 0.05, then we can conclude that they have a dependency with "Genre"
# In conclusion, Income, Age, Rental, and Incidentals have a dependency with genre.
```
